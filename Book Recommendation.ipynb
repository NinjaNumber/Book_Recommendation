# Book Recommendation

```python
import pandas as pd
import numpy as np
```

## 1. Load & Merge Datasets

```python
users = pd.read_csv('Users.csv')
users.head()
```

```python
rating = pd.read_csv('Ratings.csv')
rating.head()
```

```python
books = pd.read_csv('Books.csv')
books.head()
```

```python
users_rating = users.merge(rating, on='User-ID')
users_rating.head()
```

```python
books_rating = books.merge(rating, on='ISBN')
books_rating.head()
```

## 2. Popularity-based filtering

```python
book_count = books_rating.groupby('Book-Title')['Book-Rating'].count().reset_index().rename(columns={'Book-Rating':'n_rating'})
book_count.head()
```

```python
avg_rating = books_rating.groupby('Book-Title')['Book-Rating'].mean().reset_index().rename(columns={'Book-Rating':'Avg_ratings'})
avg_rating.head()
```

```python
df = book_count.merge(avg_rating, on='Book-Title')
df.head()
```

#### In order to implement popularity-based filtering, let us select only those books with more than 250 ratings sorted in descending order by avg_ratings.

```python
popular = df[df['n_rating']>250].sort_values(by='Avg_ratings', ascending=False).reset_index(drop=True)
popular.head()
```

## 3. Collaborative Filtering:

To build a collaborative-filtering-based recommendation system, we will consider only those users who have rated more than 200 books and those books which are having at least 50 ratings.

```python
users_rating.head()
```

```python
educator = users_rating.groupby('User-ID')['Book-Rating'].count().to_frame().rename(columns={'Book-Rating':'n_ratings'})
educator = educator[educator['n_ratings']>250].index
educator
```

```python
books_rating.head()
```

```python
books_rating = books_rating[books_rating['User-ID'].isin(educator)]
books_rating.head()
```

```python
educator_rating = books_rating.groupby('Book-Title')['Book-Rating'].count().to_frame().rename(columns={'Book-Rating':'n_rating'})
educator_rating = educator_rating[educator_rating['n_rating']>=50].index
educator_rating 
```

#### Now we have a final data frame that has only those books with at least 50 ratings and users who have rated at least 200 books.

```python
final = books_rating[books_rating['Book-Title'].isin(educator_rating)]
final.head()
```

## 4. Recommend top 5 similar books

```python
pt = final.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating').fillna(0)
pt.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_books = sorted(enumerate(similarity_scores[index]),key= lambda x: x[1], reverse =True)[1:6]
    
    for i in similar_books:
        print(pt.index[i[0]])
```

```python
recommend('4 Blondes')
```
